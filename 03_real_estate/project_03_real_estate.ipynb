## Предобработка данных

### Обработка пропусков

Прежде чем заполнить пропуски в столбце `ceiling_height`, сначала изменим значения потолков в диапазоне высоты 20-30 метров, т.к. эти данные с большой вероятностью появились в следствии ошибки при заполнении и были увеличены в 10 раз (т.е. 2,7 м. стало 27 м.):

```python
# напишем функцию, которая поделит завышенные значения высоты потолков на 10
def high_ceiling(ceiling):
    """
    Функция возвращает завышенно значение, деленное на 10 согласно правилу:
    если значение высоты потолка больше 19 и меньше 100 метров
    """
    if (ceiling > 19 and ceiling < 100):
        return ceiling / 10
    return ceiling

# и выполним ее, заменив аномальные значения
df['ceiling_height'] = df['ceiling_height'].apply(high_ceiling)
```

Исходя из того, что в Санкт-Петебурге больше исторических и премиальных  зданий с высокими (более 3м) потолками, целесообразно будет пропуски в столбце `ceiling_height` заполнить медианным значением относительно каждого отдельного населенного пункта.

```python
# заполним пропуски ceiling_height медианным значением относительно населенного пункта, где находится квартира
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('locality_name')['ceiling_height'].transform('median'))
```

После заполнения пропусков `ceiling_height`, проверим оставшееся их количество.

```python
df['ceiling_height'].isna().sum()
```

Возможно, не по всем населенным пунктам есть данные о медианных значениях. Оставшиеся пропуски заполним общей медианой по столбцу:

```python
# заполним оставшиеся пропуски ceiling_height общим медианным значением
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
```

Пропуски `floors_total` составляют лишь 0.36% от общих данных и обоснованной замены для этих прпоусков нет, поэтому целесообразно будет их удалитьж

```python
# заполним пропуски floors_total фатическим значеним этажа из столбца floor
#df['floors_total'] = df['floors_total'].fillna(df['floor'] + 1)
df = df.dropna(subset = ['floors_total']).reset_index(drop=True)
```

Для заполнения пропусков в столбцах `living_area` и `kitchen_area` рассчитаем отношение занимаемой площади жилой зоны и зоны кухни относительно общей площади исходя из строк, в которых нет пропусков. 

Два этих значения, процент занимаемой площади жилой зоной и зоной кухни, усредним и поочердно умножим на общую площадь квартиры.

```python
# пропуски living_area заполним полученными данными из отношения жилой и общей площади квартиры
df['living_area'] = df['living_area'].fillna(
    df['total_area'] * ((df['living_area'].dropna() / df['total_area']).mean()))
```

```python
# пропуски kitchen_area заполним полученными данными из отношения кухни и общей площади квартиры
df['kitchen_area'] = df['kitchen_area'].fillna(
    df['total_area'] * ((df['kitchen_area'].dropna() / df['total_area']).mean()))
```

Пропуски в столбце `is_apartment` заполним значениям False:

```python
# заполним пропуски is_apartment значениями False
df['is_apartment'] = df['is_apartment'].fillna(False)
```

Пропуски в столбце `balcony` заполним нулевым значеним, что значит отсутствие балкона в квартире:

```python
# заполним пропуски balcony значением 0
df['balcony'] = df['balcony'].fillna(0)
```

Пропусков в столбце `locality_name` очень мало, поэтому целесообразнее будет удалить данные строки, т.к. адекватной замены для этих данных нет:

```python
# удалим строки с пропусками locality_name и обновим индексацию таблицы:
df = df.dropna(subset = ['locality_name']).reset_index(drop=True)
```

Для столбца `days_exposition` адекватной замены пропусков не нашлось, но пропусков очень много (13%). Для сохранения данных проставим в пропусках 0, и учтем это  в дальнейшем.

```python
# заполним пропуски days_exposition значением 0
df['days_exposition'] = df['days_exposition'].fillna(0)
```

После обработки выведем на экран суммарное количество оставшихся пропусков:

```python
# посчитаем количество пропущенных значений по таблице после обработки
df.isna().sum()
```

### Изменение типа данных

Сначала изменим тип данных столбца `first_day_exposition`, который имеет строчный формат, но содержит в себе данные о дате публикации объявления о продаже. Данные о времени отбросим, т.к. там нулевые значения и в дальнейшем они нам не понадобятся.

```python
# изменим тип данных first_day_exposition методом to_datetime()
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d')
```

Изменим формат данных в столбцах `floors_total`, `balcony`, `airports_nearest`, `cityCenters_nearest`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`, `days_exposition`. Число с плавающей точкой (float) поменяем на целочисленное значение (int)

```python
# переводим тип данных столбцов floors_total, balcony, 
# airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, 
# ponds_around3000, ponds_nearest, days_exposition из float в int методом .astype() 
# предаврительно округлив значения методом .round()
df[['floors_total', 'balcony', 'airports_nearest', 
    'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 
    'ponds_around3000', 'ponds_nearest', 'days_exposition']] = df[
    ['floors_total', 'balcony', 'airports_nearest', 
     'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 
     'ponds_around3000', 'ponds_nearest', 'days_exposition']].round().astype('Int64')
```

Изменим формат отображения столбца `last_price` с миллиона рублей на тысячи рублей для удобства:

```python
# изменим значения столбца last_price на тысячи рублей поделив значения на 1000 и переведем в целочисленный тип данных
df['last_price'] = (df['last_price'] / 1000).round().astype('Int64')
```

Выполним изменения данных в столбце `locality_name`. Сначала изменим написание буквы "ё" на "е", для изменения слов посёлок/поселок и возможных расхождений в написании топонимов:

```python
# изменим написание "ё" на "е" методом .replace()
df['locality_name'] = df['locality_name'].str.replace('ё', 'е')
```

Добавим к топонимам "Мурино" и "Кудрово" указание "поселок" и "деревня" соответственно с помощью функции:

```python
# напишем функцию переименования "Мурино" в "поселок Мурино" и "Кудрово" в "деревню Кудрово":
def locality_rename(row):
    """
    функция возвращает название "поселок Мурино" или "деревня Кудрово", используя правило:
    в названии есть слово "Мурино" или "Кудрово" без приставки "поселок"/"деревня"
    остальные названия не трогает
    """
    if row == 'Мурино':
        return 'поселок Мурино'
    if row == 'Кудрово':
        return 'деревня Кудрово'
    return row
```

```python
# выполним вновь созданную функцию locality_rename
df['locality_name'] = df['locality_name'].apply(locality_rename)
```

## Расчёты и добавление результатов в таблицу

Для дальнейшего исследования в таблицу добавим следующие расчетные данные:
* цена квадратного метра
* день недели / месяц / год публикации объявления
* категории этажей квартиры (первый / последний / другой)
* соотношение жилой и общей площади
* соотношение кухни и общей площади

Рассчитаем цену одного квадратного метра и добавим данные в новый столбец `square_metre_price`:

```python
# рассчитаем цену за один квадратный метр поделив цену квартиры на ее площадь
# и для удобства переведем тип данных в целочисленный формат
df['square_metre_price'] = (df['last_price'] / df['total_area']).astype('int64')
```

Добавим столбцы по дате объявления: день недели, месяц, год публикации и добавим новые соответствующие столбцы:

```python
# добавим новый столбец day_exposition и заполним его данными о днях недели
df['day_exposition'] = df['first_day_exposition'].dt.dayofweek
```

```python
# добавим новый столбец month_exposition и заполним его данными о месяцах
df['month_exposition'] = df['first_day_exposition'].dt.month
```

```python
# добавим новый столбец year_exposition и заполним его данными о годах
df['year_exposition'] = df['first_day_exposition'].dt.year
```

Добавим столбец `floor_category` с категориями этажей первый / последний / другой:

```python
# напишем функцию категоризации по этажам
def floor_category(row):
    """
    Функция возвращает категорию по этажу квартиры, используя правила:
    - 'первый', если floor = 1;
    - 'последний', если floor = floors_total;
    - 'другой' для остальных этажей
    """
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    return 'другой'
```

```python
# создадим новый столбец floor_category и выполним вновь созданную функцию floor_category 
df['floor_category'] = df.apply(floor_category, axis=1)
```

```python
# для дальнейшей работы напишем функцию, которая определит id этажности квартиры
def floor_category_id(row):
    """
    Функция возвращает id по этажу квартиры, используя правила:
    - 1, если floor_category = 'первый';
    - 2, если floor_category = 'другой';
    - 3, если floor_category = 'последний'.
    """
    floor = row['floor_category']
    if floor == 'первый':
        return 1
    elif floor == 'другой':
        return 2
    return 3
```

```python
# создадим новый столбец floor_category_id и выполним вновь созданную функцию floor_category_id
df['floor_category_id'] = df.apply(floor_category_id, axis=1)
```

Рассчитаем и добавим в таблицу новые столбцы с отношениями площадей кухни и жилой зоны к общей площади квартиры:

```python
# рассчитаем соотношение жилой и общей площади, а также с отношением площади кухни к общей:
df['living_area_part'] = df['living_area'] / df['total_area']
df['kitchen_area_part'] = df['kitchen_area'] / df['total_area']
```
